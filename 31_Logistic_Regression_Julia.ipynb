# Logistic Regression - 2

```julia
using GLM
```

```julia
model = glm(@formula(camera_mp ~ price), dataset, Binomial(), LogitLink())
```

```julia
Pkg.add("DataFrames")
using DataFrames
```

```julia
predict(model, DataFrame(price=[390]))
```

```julia
predict(model, DataFrame(price=[390,400,410,420, 430, 440]))
```
