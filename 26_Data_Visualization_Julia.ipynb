# Getting started with Plots

```julia
using Pkg
```

```julia
Pkg.add("Plots")
Pkg.add("Pyplot")
Pkg.add("GR")
Pkg.add("UnicodePlots")
```

```julia
using Plots
```

```julia
gr()
```

```julia
year = [x for x in range(2000, length = 10)]
sale = 100 * rand(10)
```

```julia
plot(year, sale, labels = "Line Plot")
```

```julia
scatter!(year, sale, labels = "Scatter Plot")
```

```julia
plot(year, sale, labels = "Line Plot")
scatter!(year, sale, labels = "Scatter Plot")
```

```julia
plot(year, sale, labels = "Line Plot")
scatter!(year, sale, labels = "Scatter Plot")
xlabel!("Year")
ylabel!("Sale")
title!("Sale between 2000 and 2009")
```

```julia
x = -5:0.2:5
y = x.^2
plot(x,y)
```

# Some more Plots with Plots

```julia
p1 = plot(x,x)
p2 = plot(y,x)
p3 = plot(sin,x)
p4 = plot(cos,x)
plot(p1, p2, p3, p4)
```

```julia
p1 = plot(x,x)
p2 = plot(y,x)
p3 = plot(sin,x)
p4 = plot(cos,x)
plot(p1, p2, p3, p4, layout = (1,4))
```

```julia
plot(sin,x)
plot!(cos,x) # adicionar exclamação modifica o gráfico original
```
